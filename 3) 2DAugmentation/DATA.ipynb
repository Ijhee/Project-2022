#### 데이터 증강 프로젝트 코드를 구현하자!

    - 클러스터링 알고리즘을 사용하되 -> 평가지표(실루엣 스코어 등 존재) classification에서 사용하는 acc같은 지표를 이용
        - 그럴거면 왜 클러스터링을 하냐? 걍 classification하지??
            - 같은 범주에 속하는 데이터라고 하더라도 그 범주의 특징을 "더 잘" 설명하는 데이터와 "덜 잘" 설명하는 데이터로 구분할 수 있다.. by Clustering
            
    - 1) 5개의 알고리즘을 선정
    - 2) 클러스터링을 진행 -> acc 측정
    - 3) 클러스터 민에서 원을 그려나가서 (patient) 최대 원을 그린다
    - 4) 그 원 내부로 데이터를 생성해서 레이블을 부여한다
    
    - (차원축소 / 스케일링 / 2차원 데이터)
