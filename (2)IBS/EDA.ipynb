```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("kcy_df25.csv")
df.dropna(inplace = True)
df.reset_index(inplace = True, drop = True)
df.drop(["Unnamed: 0", "customerID"], axis = 1, inplace = True)
df
```

```python
for col in df.columns:
    if df[col].dtype == "object":
        encoder = LabelEncoder()
        df[col] = encoder.fit_transform(np.array(df[col]).reshape(-1, 1))
df.to_csv("kcy_df25.csv", index = False)
```

```python
temp = df.sample(frac = 10000 / len(df))
df = pd.concat([df, temp]).drop_duplicates(keep = False)

for col in temp.columns:
    if temp[col].dtype == "object":
        encoder = LabelEncoder()
        temp[col] = encoder.fit_transform(np.array(temp[col]).reshape(-1, 1))

for col in df.columns:
    if df[col].dtype == "object":
        encoder = LabelEncoder()
        df[col] = encoder.fit_transform(np.array(df[col]).reshape(-1, 1))

df.to_csv("kcy_df24(res).csv", index = False)
temp.to_csv("kcy_df24.csv", index = False)
```
